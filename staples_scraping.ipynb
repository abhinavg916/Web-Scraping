```
url_list = []
for i in range(1,6):
  url_list.append('https://www.staples.com/Computer-office-desks/cat_CL210795/663ea?pn=' + str(i))
```

```
item_names = []
price_list = []
```

```
# bsobj.findAll('a', {'class': 'standard-type__product_title'})[0].get_text()
```

```
# bsobj.findAll('span', {'class':'standard-type__price'})[0].get_text()
```

```
for url in url_list:
  result = requests.get(url)
  bsobj = soup(result.content,'lxml')

  product_name = bsobj.findAll('a', {'class': 'standard-type__product_title'})[0].get_text()
  product_price = bsobj.findAll('span', {'class':'standard-type__price'})[0].get_text()
  for names,price in zip(product_name,product_price):
    item_names.append(names.strip(" "))
    price_list.append(bsobj.findAll('span', {'class':'standard-type__price'})[0].get_text())
```

```
arr = []
names = 'hello world'
print(names)
arr.append(names.strings.stripped_strings)
print(arr)
```

```
import pandas as pd
df = pd.DataFrame({'Product_Name':item_names, 'Price':price_list}, columns=['Product_Name', 'Price'])
df.head()
```

```
df.to_excel('products.xlsx', index=False)
```
